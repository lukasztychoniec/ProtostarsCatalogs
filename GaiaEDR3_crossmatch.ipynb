```python
from astroquery.gaia import Gaia
Gaia.MAIN_GAIA_TABLE = "gaiaedr3.gaia_source" # Select early Data Release 3
```

```python
import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.io import ascii
import pandas as pd
```

```python
t = ascii.read('database/CatalogForGaia.tab', delimiter=' ')
t = t.to_pandas()
```

```python
def MatchGaia (gaia_catalog, coord):
    Gaia.MAIN_GAIA_TABLE = gaia_catalog # Select early Data Release 3
    width = u.Quantity(10/3600., u.deg)
    height = u.Quantity(10/3600., u.deg)
    r = Gaia.query_object_async(coordinate=coord, width=width, height=height)
    if len(r) == 0:
        match_distance = -99
        distance_pc = -99
        e_distance_pc = -99
    else:
        match_distance = r[0]['dist']*3600
        distance_pc = 1./(r[0]['parallax']/1000.)
        e_distance_pc = 1./(r[0]['parallax']/1000.)*r[0]['parallax_error']/r[0]['parallax']
    return match_distance, distance_pc, e_distance_pc
    
```

```python
for i in range(len(t)):
    coord=SkyCoord(t.loc[i, 'ra'] + ' ' + t.loc[i, 'dec'], frame='icrs', unit=(u.hourangle, u.deg))
    t.loc[i, 'GAIA_EDR3_match'], t.loc[i, 'GAIA_EDR3_dist'],t.loc[i, 'GAIA_EDR3_e_dist']= MatchGaia ("gaiaedr3.gaia_source", coord)
    t.loc[i, 'GAIA_DR2_match'], t.loc[i, 'GAIA_DR2_dist'],t.loc[i, 'GAIA_DR2_e_dist']= MatchGaia ("gaiadr2.gaia_source", coord)
```
